```python
import random
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

P=[]
for i in range (0,101):
    
    P.append(i)
print(P)
```

```python
Sample = (10,20, 50,100)
for i in Sample:
    a = random.sample(P,i)
    print(a)
    plt.hist(a, density=True, edgecolor='black', bins=30)
```
